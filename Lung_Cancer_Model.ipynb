```python
import pandas as pd 
import numpy as nmp
import matplotlib.pyplot as plt
from sklearn import linear_model as lm 
import joblib as jb
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("survey lung cancer.csv")
df
```

```python
newDF = pd.get_dummies(df.GENDER)
newDF
```

```python
updatedDF = pd.concat([df,newDF],axis='columns')
updatedDF
```

```python
X = updatedDF.drop(['GENDER','LUNG_CANCER','F'],axis = 'columns')
X
```

```python
y= df.LUNG_CANCER
y
```

```python
 X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.2)
```

```python
reg = lm.LogisticRegression( max_iter=3000)
reg.fit(X_train,y_train)
```

```python
reg.predict(X_test)
```

```python
y_pred = reg.predict(X_test)
y_pred
```

```python
reg.score(X_test,y_test)
```
